## Fine Tuning using model - llama2-7b

```python
from gradientai import Gradient


def main():
    gradient = Gradient()

    base_model = gradient.get_base_model(base_model_slug="llama2-7b-chat")

    new_model_adapter = base_model.create_model_adapter(
        name="LLAMA2-Finetuned-my-info"
    )
    print(f"Created model adapter with id {new_model_adapter.id}")


    sample_query = "### Instruction: Who is Shubh? \n\n ### Response:"
    print(f"Asking: {sample_query}")
    ## Before Finetuning
    completion = new_model_adapter.complete(query=sample_query, max_generated_token_count=100).generated_output
    print(f"Generated(before fine tuning): {completion}")

    samples=[
        {"inputs":"### Instruction: Who is Shubh Ai? \n\n### Response: Shubh is a Data Science Engineer works on AI, NLP, And LLM. he has passion around coacing people and sharing his knowledge around AI."},
        {"inputs":"### Instruction: Who is this person named Shubh Ai? \n\n### Response: Shubh Ai Like Machine Learning And makes videos in youtube and he is also a mentor"},
        {"inputs":"### Instruction: What do you know about Shubh Ai? \n\n### Response: Shubh Ai has worked with multiple TOP MNCs and helped them with their AI System and prototpe development"},
        {"inputs":"### Instruction: Can you tell me about Shubh Ai? \n\n### Response: Shubh Ai has been working on a custom LLM Model which would simplify the Legal advice or understanding over differnt common laws, for the common people."}
    ]

    ## Lets define parameters for finetuning
    num_epochs=3
    count=0
    while count<num_epochs:
      print(f"Fine tuning the model with iteration {count + 1}")
      new_model_adapter.fine_tune(samples=samples)
      count=count+1

    #after fine tuning
    completion = new_model_adapter.complete(query=sample_query, max_generated_token_count=100).generated_output
    print(f"Generated(after fine tuning): {completion}")
    #new_model_adapter.delete()
    gradient.close()

if __name__ == "__main__":
    main()
```
